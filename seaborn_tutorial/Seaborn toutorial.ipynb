```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## plotting lineplot using sns.lineplot()

```python
x = np.linspace(0 , 10 , 100)
y = np.cos(x)
plt.style.use("seaborn-darkgrid")
plt.figure(figsize=(12 , 7))
sns.lineplot(x , y )
plt.show()
```

```python
plt.figure(figsize=(14 , 6))
sns.set(rc={'axes.facecolor':'#283747',"axes.grid":True,'xtick.labelsize':14,'ytick.labelsize':14})
sns.lineplot(x , y , linewidth = 2.5, color = "#FF5722" )
sns.lineplot(x , np.tanh(y) , linewidth=2.5 , color = "#FFEB3B")
plt.show()
```

```python
sns.set(rc={'axes.facecolor':'#483457',"axes.grid":True,'xtick.labelsize':10,'ytick.labelsize':7})
plt.figure(figsize=(10,8))
sns.lineplot(x , y , color= "#FFEB3B")
plt.show()
```

## scatterplots ...

```python
plt.figure(figsize=(14 , 8))
sns.scatterplot(x =[1,4,7,8,5,4,7,8,9,5] , y = [1,7,4,5,2,8,0,9,6,3] , color="#FF0100" )
sns.scatterplot(np.linspace(1 , 4 , 6) , np.linspace(2 , 5 , 6) , color="#457896" , marker=">" )
plt.legend("best")
plt.show()
```

## bar plot ...

```python
plt.figure(figsize=(10 , 6))
sns.barplot(x =[1,4,7,8,5,4,7,8,9,5] , y = [1,7,4,5,2,8,0,9,6,3])
plt.show()
```

```python
plt.figure(figsize=(7 , 10))
sns.set(rc={"axes.facecolor":"#283747", "axes.grid":False,'xtick.labelsize':14,'ytick.labelsize':14})
sns.barplot([1,4,7,8,5,4,7,8,9,5] ,[1,7,4,5,2,8,0,9,6,3])
plt.show()
```

## count plot ...

## dist plot ..

```python

plt.figure(figsize=(13 , 7))
sns.displot([4,5,7,0,8,9,6,5,8,4,5,6,2,3,2,5] , kde = True , rug=True)
plt.show()
```
